```python
import pandas as pd
```

```python
prophet=pd.read_csv("PROPHET_CA_2_final.csv")
prophet=prophet[["ds","result"]][-28:].T
prophet.columns=["F{}".format(x) for x in range(1,29)]
prophet.drop("ds",axis=0,inplace =True)
prophet=prophet.reset_index()
prophet.rename(columns={"index":"id"},inplace =True)
prophet
```

```python
fold=r"model3_3fold.csv"
```

```python
DATA=pd.read_csv(fold)
DATA
```

```python
DATA=DATA[DATA["id"].str.contains("CA_2_evaluation")].reset_index(drop=True)
DATA
```

```python
scaling=pd.DataFrame()
for i in range(1,29):
    scaling["F{}".format(i)]=DATA["F{}".format(i)]/DATA["F{}".format(i)].sum()
scaling
```

```python
DATA_copy=DATA.copy()
for i in range(1,29):
    DATA_copy["F{}".format(i)]=scaling["F{}".format(i)]*prophet["F{}".format(i)].values
DATA_copy
```

```python
SUB=pd.read_csv(fold)
for item in SUB:
    if item!='id':
        SUB[item].loc[(SUB['id'].isin(DATA_copy['id']))]=list(DATA_copy[item])
```
