```python
import pandas as pd
```

```python
cluster_0_fold0=pd.read_csv('all_preds_cluster_00.csv')
cluster_0_fold1=pd.read_csv('all_preds_cluster_01.csv')
cluster_0_fold2=pd.read_csv('all_preds_cluster_02.csv')
```

```python
cluster_0_fold0.head()
```

```python
cluster_0_fold0.shape
```

```python
cluster_0=pd.DataFrame()
cluster_0['id']=cluster_0_fold0['id']
for item in cluster_0_fold0:
    if item!='id':
        cluster_0[item]=(cluster_0_fold0[item]*(1/3))+(cluster_0_fold1[item]*(1/3))+(cluster_0_fold2[item]*(1/3))
```

```python
cluster_0.head()
```

```python
cluster_0.shape
```

```python
sub=pd.read_csv('lgb_bycluster_cluster4_noweather_v6.csv')
```

```python
for item in sub:
    if item!='id':
        sub[item].loc[(sub['id'].isin(cluster_0['id']))]=list(cluster_0[item])
```
