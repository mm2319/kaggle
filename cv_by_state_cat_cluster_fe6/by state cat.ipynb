```python
cafoods_fold0=pd.read_csv('all_preds_CAFOODS_0.csv')
cafoods_fold1=pd.read_csv('all_preds_CAFOODS_1.csv')
cafoods_fold2=pd.read_csv('all_preds_CAFOODS_2.csv')
```

```python
cafoods_fold0.head()
```

```python
cafoods_fold0.shape
```

```python
cafoods=pd.DataFrame()
cafoods['id']=cafoods_fold0['id']
for item in cafoods_fold0:
    if item!='id':
        cafoods[item]=(cafoods_fold0[item]*(1/3))+(cafoods_fold1[item]*(1/3))+(cafoods_fold2[item]*(1/3))
```

```python
cafoods.head()
```

```python
sub=pd.read_csv('lgb_bystcat_clusterfe6.csv')
```

```python
for item in sub:
    if item!='id':
        sub[item].loc[(sub['id'].isin(cafoods['id']))]=list(cafoods[item])
```

```python
sub.to_csv('lgb_bystcat_clusterfe6_cafoods.csv',index=False)
```
